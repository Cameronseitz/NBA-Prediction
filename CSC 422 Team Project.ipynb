```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
def k_fold_accuracy(k, model, X_data, Y_data):
    """
    You will need to implement a k-fold cross validation with accuracy as the performance metric.
    
    Your inputs and outputs are specified as below:
    
    Input:
        k: An integer indicates the number of fold to do cross validation.
        model: A DecisionTreeClassifier instance.
        X_data: A numpy array of shape (n_data_rows, n_attributes) where n_data_rows refers to 
              the number of rows in your dataset and n_attributes refers to the number of attributes. 
              This typically is from the training dataset.
        Y_data: A numpy array of shape (n_data_rows, ) containing the class labels for each row in your 
              dataset. This typically is from the training dataset.
    Output:
        scores: A list of length k, with scores[i] containing the validation accuracy score for the i-th fold.
    """
    
    # Init k-fold splitter
    kf = KFold(n_splits=k)
    rsmes = [0] * k
    i=0
    for train_index, test_index in kf.split(X_data):
        
        X_train = X_data.drop(X_data.index[test_index])
        Y_train = Y_data.drop(Y_data.index[test_index])
        
        X_test = X_data.drop(X_data.index[train_index])
        Y_test = Y_data.drop(Y_data.index[train_index])
        
        model.fit(X=X_train, y=Y_train)
        predictions = model.predict(X_test)
        mse = mean_squared_error(Y_test, predictions)
        rsme = math.sqrt(mse)
        rsmes[i] = rsme
        i=i+1
        
        score = np.mean(rsmes)

    return score
```

# Using data from all four eras

## Prepare the data

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

#no longer should be considered
df = df.drop(['teamstatspk', 'SEASON', 'REB', '+/-'], axis=1)

# scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
# create a new DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

# features are everything but the teamstatspk, season, rebounds, and target
X = scaled_df.drop(['WIN%'], axis=1)

# target is win %
y = scaled_df['WIN%']

# split the data into training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Find the best alpha for Lasso

```python
#iterate through an array of alphas to find the best
alphas = np.arange(0.0001, 0.0007, 0.00001)
model_scores = [0] * len(alphas)
test_model = [0] * len(alphas)
test_rmse = [0] * len(alphas)
for i in range(len(alphas)):
    test_model[i] = Lasso(alpha=alphas[i])
    
    
    model_scores[i] = k_fold_accuracy(10, test_model[i], X_train, y_train)
    

    
plt.plot(alphas, model_scores, color='red')

plt.xlabel("alpha value")
plt.ylabel(f'Average RSME of 10-fold validation')
```

## Create model with best alpha

```python

# create and train linear regression model
lr = Lasso(alpha=0.0001)
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)

# accuracies
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print("mse: " + str(mse))
print("r2: " + str(r2))


# most important coefficients
coefficients = pd.DataFrame({'feature': X.columns, 'coefficient': lr.coef_})
coefficients.sort_values(by='coefficient', ascending=False, inplace=True)
print(coefficients)

rsme = k_fold_accuracy(10, lr, X, y)
print("rsme: " + str(rsme))

```

# Create Lasso model with second degree variables

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
df_squared = df

df_squared['PTS^2'] = df['PTS'].pow(2)
df_squared['FGM^2'] = df['FGM'].pow(2)
df_squared['FGA^2'] = df['FGA'].pow(2)
df_squared['FG%^2'] = df['FG%'].pow(2)
df_squared['3PA^2'] = df['3PA'].pow(2)
df_squared['3P%^2'] = df['3P%'].pow(2)
df_squared['FTM^2'] = df['FTM'].pow(2)
df_squared['FTA^2'] = df['FTA'].pow(2)
df_squared['FT%^2'] = df['FT%'].pow(2)
df_squared['OREB^2'] = df['OREB'].pow(2)
df_squared['DREB^2'] = df['DREB'].pow(2)
df_squared['AST^2'] = df['AST'].pow(2)
df_squared['TOV^2'] = df['TOV'].pow(2)
df_squared['STL^2'] = df['STL'].pow(2)
df_squared['BLK^2'] = df['BLK'].pow(2)
df_squared['BLKA^2'] = df['BLKA'].pow(2)
df_squared['PF^2'] = df['PF'].pow(2)
df_squared['PFD^2'] = df['PFD'].pow(2)
# df_squared['+/-^2'] = df['+/-'].pow(2)



scaler = MinMaxScaler()
scaled_data_squared = scaler.fit_transform(df_squared)
# create a new DataFrame with the scaled data
scaled_df_squared = pd.DataFrame(scaled_data_squared, columns=df_squared.columns)

# features are everything but the teamstatspk, season, rebounds, and target
X = scaled_df_squared.drop(['WIN%'], axis=1)


# target is win %
y = scaled_df_squared['WIN%']

# split the data into training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# create and train linear regression model
# lr_squared = Lasso(alpha=0.0003)
lr_squared = Lasso(alpha=0.0001, max_iter=5000)
lr_squared.fit(X_train, y_train)
predictions = lr_squared.predict(X_test)

# accuracies
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print("mse: " + str(mse))
print("r2: " + str(r2))

rsme = k_fold_accuracy(10, lr_squared, X, y)
print("rsme: " + str(rsme))
print()


# most important coefficients
coefficients = pd.DataFrame({'feature': X.columns, 'coefficient': lr_squared.coef_})
coefficients.sort_values(by='coefficient', ascending=False, inplace=True)
print(coefficients)





```

# Each Era - Lasso Linear Regression

```python
df_eras = [df_1, df_2, df_3, df_4]

# initialize a dictionary to store the coefficients for each feature across eras
coefficients_dict = {}
for feature in X.columns:
    coefficients_dict[feature] = []
    
for i, era in enumerate(df_eras):
    print("Era: " + str(i + 1))
    #no longer should be considered
    era = era.drop(['teamstatspk', 'SEASON', 'REB', '+/-'], axis=1)

    #scal the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(era)
    # create a new DataFrame with the scaled data
    scaled_era = pd.DataFrame(scaled_data, columns=era.columns)

    # features are everything but the teamstatspk, season, and target
    X = scaled_era.drop(['WIN%'], axis=1)

    # target is win %
    y = scaled_era['WIN%']

    # split the data into training and test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # create and train linear regression model
    lr = Lasso(alpha=0.0001)
    lr.fit(X_train, y_train)
    predictions = lr.predict(X_test)

    # accuracies
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    print("mse: " + str(mse))
    print("r2: " + str(r2))
    
    rsme = k_fold_accuracy(10, lr, X, y)
    print("rsme: " + str(rsme))
    print()
    
    # store the coefficients for each feature in the dictionary
    for feature, coef in zip(X.columns, lr.coef_):
        coefficients_dict[feature].append(coef)

    # print coefficients
    coefficients = pd.DataFrame({'feature': X.columns, 'coefficient': lr.coef_})
    coefficients.sort_values(by='coefficient', ascending=False, inplace=True)
    print(coefficients)

    print()
```

```python
# plot all the features on one graph with different colors
fig, ax = plt.subplots()
for i, feature in enumerate(X.columns):
    ax.plot(range(1, 5), coefficients_dict[feature], label=feature)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks([1, 2, 3, 4])
ax.set_xlabel('Era')
ax.set_ylabel('Coefficient Value')
plt.show()
```

```python
# calculate the average absolute value for each feature
avg_abs_coef = {feature: np.mean(np.abs(coefficients_dict[feature])) for feature in coefficients_dict}

# sort the features based on the average absolute value
sorted_features = sorted(avg_abs_coef, key=avg_abs_coef.get, reverse=True)

# select the top 10 features
top_10_features = sorted_features[:10]

# plot the top 10 features with custom colors and labels
fig, ax = plt.subplots()

for i, feature in enumerate(top_10_features):
    ax.plot(range(1, 5), coefficients_dict[feature], label=feature)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xticks([1, 2, 3, 4])
ax.set_xlabel('Era')
ax.set_ylabel('Coefficient Value')
plt.show()
```
